```python
import pandas as pd
import matplotlib.pyplot as plt


# Function to calculate average heFG and aeFG for a given team
def calculate_team_efg(team):
    home_games = data[data['Home'] == team]
    away_games = data[data['Away'] == team]
    avg_heFG_home = home_games['heFG'].mean()
    avg_aeFG_away = away_games['aeFG'].mean()
    avg_eFG = (avg_heFG_home + avg_aeFG_away) / 2
    return avg_heFG_home, avg_aeFG_away, avg_eFG

# Calculate averages for Carleton, Queen's, and Waterloo
avg_heFG_home_carleton, avg_aeFG_away_carleton, avg_eFG_carleton = calculate_team_efg('Carleton')
avg_heFG_home_queens, avg_aeFG_away_queens, avg_eFG_queens = calculate_team_efg("Queen's")
avg_heFG_home_waterloo, avg_aeFG_away_waterloo, avg_eFG_waterloo = calculate_team_efg('Waterloo')

# Calculate averages for the rest of the league
non_carleton_queens_waterloo_home = data[~data['Home'].isin(['Carleton', "Queen's", 'Waterloo'])]
non_carleton_queens_waterloo_away = data[~data['Away'].isin(['Carleton', "Queen's", 'Waterloo'])]
avg_heFG_home_rest = non_carleton_queens_waterloo_home['heFG'].mean()
avg_aeFG_away_rest = non_carleton_queens_waterloo_away['aeFG'].mean()
avg_eFG_rest = (avg_heFG_home_rest + avg_aeFG_away_rest) / 2

# Print the results
print(f"Carleton Home heFG: {avg_heFG_home_carleton}")
print(f"Carleton Away aeFG: {avg_aeFG_away_carleton}")
print(f"Carleton Overall eFG: {avg_eFG_carleton}")

print(f"Queen's Home heFG: {avg_heFG_home_queens}")
print(f"Queen's Away aeFG: {avg_aeFG_away_queens}")
print(f"Queen's Overall eFG: {avg_eFG_queens}")

print(f"Waterloo Home heFG: {avg_heFG_home_waterloo}")
print(f"Waterloo Away aeFG: {avg_aeFG_away_waterloo}")
print(f"Waterloo Overall eFG: {avg_eFG_waterloo}")

print(f"Rest of League Home heFG: {avg_heFG_home_rest}")
print(f"Rest of League Away aeFG: {avg_aeFG_away_rest}")
print(f"Rest of League Overall eFG: {avg_eFG_rest}")

# Plotting the results
teams = ['Carleton', "Queen's", 'Waterloo', 'Rest of League']
colors = ['red', 'blue', 'gold', 'grey']
avg_eFG = [avg_eFG_carleton, avg_eFG_queens, avg_eFG_waterloo, avg_eFG_rest]

x = range(len(teams))

plt.figure(figsize=(10, 6))
bars = plt.bar(x, avg_eFG, width=0.2, color=colors, alpha=1, align='center')

# Adding text annotations
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.005, round(yval, 2), va='bottom', ha='center', fontsize=9, color='black')

# Adjusting y-axis limits
plt.ylim(0, max(avg_eFG) + 0.1)

plt.xlabel('Teams')
plt.ylabel('Average eFG')
plt.title('Average Effective Field Goal Percentages (eFG)')
plt.xticks(x, teams)
plt.legend()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Function to calculate average eFG for a given team
def calculate_team_efg(team):
    home_games = data[data['Home'] == team]
    away_games = data[data['Away'] == team]
    avg_heFG_home = home_games['heFG'].mean()
    avg_aeFG_away = away_games['aeFG'].mean()
    avg_eFG = (avg_heFG_home + avg_aeFG_away) / 2
    return avg_heFG_home, avg_aeFG_away, avg_eFG

# Function to calculate average FTF for a given team
def calculate_team_FTF(team):
    home_games = data[data['Home'] == team]
    away_games = data[data['Away'] == team]
    avg_hFTF_home = home_games['hFTF'].mean()
    avg_aFTF_away = away_games['aFTF'].mean()
    return (avg_hFTF_home + avg_aFTF_away) / 2

# Function to calculate average TOV for a given team
def calculate_team_TOV(team):
    home_games = data[data['Home'] == team]
    away_games = data[data['Away'] == team]
    avg_hTOV_home = home_games['hTOV'].mean()
    avg_aTOV_away = away_games['aTOV'].mean()
    return (avg_hTOV_home + avg_aTOV_away) / 2

# Teams to analyze
teams = ['Carleton', "Queen's", 'Waterloo']
colors = ['red', 'blue', 'gold']

# Calculate averages for each team
eFG_values = [calculate_team_efg(team)[2] for team in teams]
FTF_values = [calculate_team_FTF(team) for team in teams]
TOV_values = [calculate_team_TOV(team) for team in teams]

# Calculate averages for the rest of the league
avg_eFG_rest = (data[~data['Home'].isin(teams)]['heFG'].mean() + data[~data['Away'].isin(teams)]['aeFG'].mean()) / 2
avg_FTF_rest = (data[~data['Home'].isin(teams)]['hFTF'].mean() + data[~data['Away'].isin(teams)]['aFTF'].mean()) / 2
avg_TOV_rest = (data[~data['Home'].isin(teams)]['hTOV'].mean() + data[~data['Away'].isin(teams)]['aTOV'].mean()) / 2

# Set up the bar positions
x = range(len(teams) * 3)  # 3 statistics for each team
width = 0.2  # Width of the bars

# Create the figure and axes
fig, ax = plt.subplots(figsize=(20, 12))

# Plotting the values for each statistic for each team
for i, team in enumerate(teams):
    ax.bar([pos + i * width for pos in range(3)], 
           [eFG_values[i], FTF_values[i], TOV_values[i]], 
           width, label=team, color=colors[i])

# Adding rest of league averages
ax.bar([pos + len(teams) * width for pos in range(3)], 
       [avg_eFG_rest, avg_FTF_rest, avg_TOV_rest], 
       width, label='Rest of League', color='grey', alpha=0.7)

# Adding labels, title, and custom x-axis tick labels
ax.set_xlabel('Statistics')
ax.set_ylabel('Averages')
ax.set_title('Average Statistics Comparison')
ax.set_xticks([pos + (len(teams) - 1) * width / 2 for pos in range(3)])
ax.set_xticklabels(['eFG', 'FTF', 'TOV'])
ax.legend(title='Teams')

# Adjust y-axis limits for better visibility
ax.set_ylim(0, max(max(eFG_values), max(FTF_values), max(TOV_values), avg_eFG_rest, avg_FTF_rest, avg_TOV_rest) + 0.1)

# Function to add labels to the bars
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # Vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

# Labeling bars
autolabel(ax.patches)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Filter data to only include the 2023-2024 season
data_2023_2024 = data[data['Year'] == '2023-2024']

# Function to calculate average stats for a given team
def calculate_team_stats(team):
    home_games = data_2023_2024[data_2023_2024['Home'] == team]
    away_games = data_2023_2024[data_2023_2024['Away'] == team]
    
    avg_heFG_home = home_games['heFG'].mean()
    avg_aeFG_away = away_games['aeFG'].mean()
    avg_eFG = (avg_heFG_home + avg_aeFG_away) / 2
    
    avg_hFTF_home = home_games['hFTF'].mean()
    avg_aFTF_away = away_games['aFTF'].mean()
    avg_FTF = (avg_hFTF_home + avg_aFTF_away) / 2
    
    avg_hTOV_home = home_games['hTOV'].mean()
    avg_aTOV_away = away_games['aTOV'].mean()
    avg_TOV = (avg_hTOV_home + avg_aTOV_away) / 2
    
    avg_hORB_home = home_games['hORB'].mean()
    avg_aORB_away = away_games['aORB'].mean()
    avg_ORB = (avg_hORB_home + avg_aORB_away) / 2
    
    avg_hDRB_home = home_games['hDRB'].mean()
    avg_aDRB_away = away_games['aDRB'].mean()
    avg_DRB = (avg_hDRB_home + avg_aDRB_away) / 2

    return avg_eFG, avg_FTF, avg_TOV, avg_ORB, avg_DRB

# Teams to analyze
teams = ['Carleton', "Queen's", 'Waterloo']
colors = ['red', 'blue', 'gold']

# Calculate averages for each team
team_stats = {team: calculate_team_stats(team) for team in teams}

# Calculate averages for the rest of the league
rest_of_league_home = data_2023_2024[~data_2023_2024['Home'].isin(teams)]
rest_of_league_away = data_2023_2024[~data_2023_2024['Away'].isin(teams)]

avg_eFG_rest = (rest_of_league_home['heFG'].mean() + rest_of_league_away['aeFG'].mean()) / 2
avg_FTF_rest = (rest_of_league_home['hFTF'].mean() + rest_of_league_away['aFTF'].mean()) / 2
avg_TOV_rest = (rest_of_league_home['hTOV'].mean() + rest_of_league_away['aTOV'].mean()) / 2
avg_ORB_rest = (rest_of_league_home['hORB'].mean() + rest_of_league_away['aORB'].mean()) / 2
avg_DRB_rest = (rest_of_league_home['hDRB'].mean() + rest_of_league_away['aDRB'].mean()) / 2

# Set up the bar positions
x = range(5)  # 5 statistics to plot
width = 0.2  # Width of the bars

# Create the figure and axes
fig, ax = plt.subplots(figsize=(20, 12))

# Plotting the values for each statistic for each team
for i, team in enumerate(teams):
    stats = team_stats[team]
    bars = ax.bar([pos + i * width for pos in x], stats, width, label=team, color=colors[i])
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # Vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

# Adding rest of league averages
rest_stats = [avg_eFG_rest, avg_FTF_rest, avg_TOV_rest, avg_ORB_rest, avg_DRB_rest]
bars = ax.bar([pos + len(teams) * width for pos in x], rest_stats, width, label='Rest of League', color='grey', alpha=0.7)
for bar in bars:
    height = bar.get_height()
    ax.annotate(f'{height:.2f}',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3),  # Vertical offset
                textcoords="offset points",
                ha='center', va='bottom')

# Adding labels, title, and custom x-axis tick labels
ax.set_xlabel('Statistics')
ax.set_ylabel('Averages')
ax.set_title('Average Statistics Comparison (2023-2024 Season)')
ax.set_xticks([pos + (len(teams) - 1) * width / 2 for pos in x])
ax.set_xticklabels(['eFG', 'FTF', 'TOV', 'ORB', 'DRB'])
ax.legend(title='Teams')

# Adjust y-axis limits for better visibility
ax.set_ylim(0, max(max(stats) for stats in list(team_stats.values()) + [rest_stats]) + 0.1)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Filter data to only include the 2023-2024 season
data_2023_2024 = data[data['Year'] == '2023-2024']

# Filter data for games Waterloo played
waterloo_games = data_2023_2024[(data_2023_2024['Home'] == 'Waterloo') | (data_2023_2024['Away'] == 'Waterloo')]

# Determine the outcome for Waterloo in each game
waterloo_games['Outcome'] = waterloo_games.apply(
    lambda row: 'Win' if (row['Home'] == 'Waterloo' and row['HTWon'] == 1) or 
                            (row['Away'] == 'Waterloo' and row['HTWon'] == 0) else 'Loss', axis=1)

# Iterate through each game and create a bar chart
for index, game in waterloo_games.iterrows():
    fig, ax = plt.subplots(figsize=(12, 8))
    
    stats = ['eFG', 'TOV', 'ORB', 'DRB', 'FTF']
    home_stats = [round(game['heFG'], 2), round(game['hTOV'], 2), round(game['hORB'], 2), round(game['hDRB'], 2), round(game['hFTF'], 2)]
    away_stats = [round(game['aeFG'], 2), round(game['aTOV'], 2), round(game['aORB'], 2), round(game['aDRB'], 2), round(game['aFTF'], 2)]
    
    x = range(len(stats))
    
    ax.bar([pos - 0.2 for pos in x], home_stats, width=0.4, label=game['Home'], color='yellow' if game['Home'] == 'Waterloo' else 'red')
    ax.bar([pos + 0.2 for pos in x], away_stats, width=0.4, label=game['Away'], color='yellow' if game['Away'] == 'Waterloo' else 'red')
    
    ax.set_xlabel('Statistics')
    ax.set_ylabel('Value')
    ax.set_title(f"Game {index+1}: {game['Home']} ({game['hPoints']}) vs {game['Away']} ({game['aPoints']}) - {game['Outcome']}")
    ax.set_xticks(x)
    ax.set_xticklabels(stats)
    ax.legend()
    
    for i, v in enumerate(home_stats):
        ax.text(i - 0.2, v + 0.02, f"{v:.2f}", ha='center')
    
    for i, v in enumerate(away_stats):
        ax.text(i + 0.2, v + 0.02, f"{v:.2f}", ha='center')
    
    plt.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

# Load the data
data = pd.read_csv('Historical Data 2015-2023.csv')

# Filter the data for the 2023-2024 season where Waterloo is away
data_filtered = data[(data['Year'] == '2023-2024') & 
                     (data['Away'] == 'Waterloo')]

# Add a column for Outcome based on HTWon
# Since Waterloo is the away team, HTWon=0 means Waterloo won
data_filtered['Outcome'] = data_filtered['HTWon'].apply(lambda x: 'Win' if x == 0 else 'Loss')

# Melt the data to long format for statistics
data_long = data_filtered.melt(id_vars=['Home', 'Away', 'Outcome', 'Year'], 
                               value_vars=['heFG', 'aeFG', 'hTOV', 'aTOV', 'hORB', 'aORB', 'hDRB', 'aDRB', 'hFTF', 'aFTF'],
                               var_name='Statistic', value_name='Value')

# Create a new column to differentiate between Home and Away statistics
data_long['Team'] = data_long['Statistic'].apply(lambda x: 'Waterloo' if x.startswith('a') else 'Opponent')

# Define color mapping
color_map = {'Waterloo': 'yellow', 'Opponent': 'red'}

# Create separate plots for each game
figures = []
for game_date, game_data in data_long.groupby('Year'):
    fig = go.Figure()

    # Plot Waterloo statistics
    fig.add_trace(go.Bar(
        x=game_data[game_data['Team'] == 'Waterloo']['Statistic'],
        y=game_data[game_data['Team'] == 'Waterloo']['Value'],
        name='Waterloo',
        marker_color='yellow'
    ))

    # Plot Opponent statistics
    fig.add_trace(go.Bar(
        x=game_data[game_data['Team'] == 'Opponent']['Statistic'],
        y=game_data[game_data['Team'] == 'Opponent']['Value'],
        name='Opponent',
        marker_color='red'
    ))

    # Update layout for each game plot
    fig.update_layout(
        title=f"Game on {game_date} (Outcome: {data_filtered[data_filtered['Year'] == game_date]['Outcome'].values[0]})",
        xaxis_title='Statistic',
        yaxis_title='Statistic Value',
        barmode='group',
        height=400,
        width=600
    )
    
    figures.append(fig)

# Show all figures
for fig in figures:
    fig.show()
```
