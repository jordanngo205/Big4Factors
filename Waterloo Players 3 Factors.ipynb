```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Calculate the number of distinct games
num_games = df['Opponent'].nunique()

# Group by Player and calculate the total and average FGA per game
total_fga_per_player = df.groupby('Player')['FGA'].sum().reset_index()
total_fga_per_player['Average_FGA'] = total_fga_per_player['FGA'] / num_games

# Sort the players by the highest average FGA
total_fga_per_player = total_fga_per_player.sort_values(by='Average_FGA', ascending=False)

# Calculate the total FGA for the entire team per game
total_fga_per_game = df['FGA'].sum() / num_games

# Plotting the average Field Goals Attempted (FGA) per player
plt.figure(figsize=(16, 8))
barplot = sns.barplot(x='Player', y='Average_FGA', data=total_fga_per_player)

# Setting the title and labels with bigger font sizes
plt.title(f'Average Field Goals Attempted (FGA) per Player in the Entire Season\nAverage Total Team FGA per Game: {total_fga_per_game:.1f}', fontsize=20)
plt.xlabel('Player', fontsize=16)
plt.ylabel('Average FGA', fontsize=16)
plt.xticks(rotation=45, fontsize=14)
plt.yticks(fontsize=14)

# Annotate the values on top of the bars
for p in barplot.patches:
    barplot.annotate(format(p.get_height(), '.1f'),
                     (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha='center', va='center',
                     xytext=(0, 9),
                     textcoords='offset points',
                     fontsize=14)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the number of distinct games
num_games = df['Opponent'].nunique()

# Group by Player and calculate the total and average FGA per game
total_fga_per_player = df.groupby('Player')['FGA'].sum().reset_index()
total_fga_per_player['Average_FGA'] = total_fga_per_player['FGA'] / num_games

# Calculate the average eFG% per player
average_efg_per_player = df.groupby('Player')['eFG'].mean().reset_index()

# Merge the FGA and eFG% dataframes
player_stats = pd.merge(total_fga_per_player, average_efg_per_player, on='Player')

# Sort the players by the highest average FGA
player_stats = player_stats.sort_values(by='Average_FGA', ascending=False)

# Calculate the total FGA for the entire team per game
total_fga_per_game = df['FGA'].sum() / num_games

# Plotting the average Field Goals Attempted (FGA) and eFG% per player
fig, ax1 = plt.subplots(figsize=(18, 10))

# Bar plot for Average FGA
color = 'tab:blue'
ax1.set_xlabel('Player', fontsize=16)
ax1.set_ylabel('Average FGA', fontsize=16, color=color)
barplot = sns.barplot(x='Player', y='Average_FGA', data=player_stats, ax=ax1, color=color)
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45, fontsize=14)
ax1.set_ylim(0, player_stats['Average_FGA'].max() + 5)  # Adjusted y-axis limit

# Annotate the values on top of the bars
for p in barplot.patches:
    barplot.annotate(format(p.get_height(), '.1f'),
                     (p.get_x() + p.get_width() / 2., p.get_height() + 0.3),
                     ha='center', va='bottom',
                     fontsize=14, color=color)

# Title
ax1.set_title(f'Average Field Goals Attempted (FGA) and Effective Field Goal Percentage (eFG%)\nper Player in the Entire Season\nAverage Total Team FGA per Game: {total_fga_per_game:.1f}', fontsize=20)

# Create a second y-axis for the eFG% line plot
ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Average eFG%', fontsize=16, color=color)
lineplot = sns.lineplot(x='Player', y='eFG', data=player_stats, ax=ax2, color=color, marker='o', linewidth=2, markersize=8)
ax2.tick_params(axis='y', labelcolor=color)
ax2.set_ylim(0, max(player_stats['eFG']) + 0.2)  # Increased y-axis limit

# Annotate the eFG% values on the line plot with increased vertical offset
for x, y in zip(lineplot.get_lines()[0].get_xdata(), lineplot.get_lines()[0].get_ydata()):
    ax2.text(x, y + 0.01, f'{y:.2f}', color=color, ha='center', va='bottom', fontsize=14)  # Increased offset

# Adjust layout to prevent overlap
fig.tight_layout()  
plt.show()
```

```python
import pandas as pd

# Load the player's stats data
players_stats = pd.read_csv('Waterloo Players Stats 23-24.csv')

# Remove specific players
players_to_remove = ['Summer Pahl', 'Sonja Matthews', 'Aimee Boulay', 'Jemima Lacroix', 'Ghiselle Poblete', 'Alliasen Miscione']
players_stats = players_stats[~players_stats['Player'].isin(players_to_remove)]

# Add Kymora Stafford to the dataframe with her eFG% and FGA
new_player = pd.DataFrame({'Player': ['Kymora Stafford'], 'eFG': [0.436], 'FGA': [8.928571]})
players_stats = pd.concat([players_stats, new_player], ignore_index=True)

# Calculate the total effective field goal percentage (eFG%) for each player
players_stats_grouped = players_stats.groupby('Player').agg({'FGA': 'mean', 'eFG': 'mean'}).reset_index()

# Calculate each player's proportion of the total eFG%
total_team_efg = (players_stats_grouped['FGA'] * players_stats_grouped['eFG']).sum()
players_stats_grouped['eFG Proportion'] = (players_stats_grouped['FGA'] * players_stats_grouped['eFG']) / total_team_efg

# Adjust the shot allocation based on these proportions, scaled to match the new minimum of 65 shots per game
players_stats_grouped['Adjusted FGA Next Season'] = players_stats_grouped['eFG Proportion'] * 65

# Set specific FGAs as per your requirement
players_stats_grouped.loc[players_stats_grouped['Player'] == 'Yemi Oladipo', 'Adjusted FGA Next Season'] = 4
players_stats_grouped.loc[players_stats_grouped['Player'] == 'Tamara Popovic', 'Adjusted FGA Next Season'] = 3
players_stats_grouped.loc[players_stats_grouped['Player'] == 'Mya Jickling', 'Adjusted FGA Next Season'] = 0



# Add new player Kate with the same adjusted FGA value as half of Kymora Stafford's adjusted FGA
new_player_kate = pd.DataFrame({'Player': ['Kate Huh'], 'eFG': [0], 'Adjusted FGA Next Season': [3]})
players_stats_grouped = pd.concat([players_stats_grouped, new_player_kate], ignore_index=True)
players_stats_grouped.loc[players_stats_grouped['Player'] == 'Mya Jickling', 'Adjusted FGA Next Season'] = 3

# Calculate the total FGAs set explicitly
explicit_fg_total = players_stats_grouped.loc[players_stats_grouped['Player'].isin(['Yemi Oladipo', 'Tamara Popovic', 'Mya Jickling', 'Kate Huh']), 'Adjusted FGA Next Season'].sum()

# Calculate the remaining FGAs to be distributed
remaining_fga = 65 - explicit_fg_total

# Calculate the new proportions for the remaining players
remaining_players = players_stats_grouped[~players_stats_grouped['Player'].isin(['Yemi Oladipo', 'Tamara Popovic', 'Mya Jickling', 'Kate Huh'])]
remaining_players['Adjusted FGA Next Season'] *= remaining_fga / remaining_players['Adjusted FGA Next Season'].sum()

# Update the adjusted FGA for remaining players
players_stats_grouped.update(remaining_players)

# Reduce one shot from Kymora and Kate, and add these shots to Nehita Oko-Oboh and Jaime Newell

players_stats_grouped.loc[players_stats_grouped['Player'] == 'Kate Huh', 'Adjusted FGA Next Season'] -= 1

# Add one shot each to Nehita Oko-Oboh and Jaime Newell
for player in ['Nehita Oko-Oboh', 'Jaime Newell']:
    if player in players_stats_grouped['Player'].values:
        players_stats_grouped.loc[players_stats_grouped['Player'] == player, 'Adjusted FGA Next Season'] += 1
    else:
        new_player = pd.DataFrame({'Player': [player], 'eFG': [0], 'Adjusted FGA Next Season': [1]})
        players_stats_grouped = pd.concat([players_stats_grouped, new_player], ignore_index=True)

# Display the results
print(players_stats_grouped[['Player', 'FGA', 'eFG', 'Adjusted FGA Next Season']])
```
